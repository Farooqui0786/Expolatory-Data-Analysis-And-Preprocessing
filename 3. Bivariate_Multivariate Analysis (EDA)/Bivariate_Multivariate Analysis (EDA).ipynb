```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')     # restaurant tip dataset
```

```python
tips.head()
```

```python
titanic = pd.read_csv('./train.csv')      # titanic dataset
```

```python
titanic.head()
```

```python
flights = pd.read_csv('./flights.csv')      # flight dataset
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## **1. Scatterplot (Numerical - Numerical)**

```python
# bivariate
sns.scatterplot(tips, x="total_bill", y="tip")
```

```python
# multivariate
plt.figure(figsize=(10,8))
sns.scatterplot(tips, x="total_bill", y="tip", hue="sex")
```

```python
# multivariate
plt.figure(figsize=(10,8))
sns.scatterplot(tips, x="total_bill", y="tip", hue="sex", style="smoker")
```

```python
# multivariate
plt.figure(figsize=(10,8))
sns.scatterplot(tips, x="total_bill", y="tip", hue="sex", style="smoker", size="size")
```

## **2. Barplot (Numerical - Categorical)**

```python
# bivariate
sns.barplot(titanic, x="Pclass", y="Age")
```

```python
# multivariate
sns.barplot(titanic, x="Pclass", y="Fare", hue="Sex")
```

## **3. Boxplot (Numerical - Categorical)**

```python
# bivariate
sns.boxplot(titanic, x="Sex", y="Age")
```

```python
# bivariate
sns.boxplot(titanic, x="Sex", y="Age", hue="Survived")
```

## **4. Distplot (Numerical - Categorical)**

```python
sns.histplot(titanic, x="Age", hue="Survived", kde=True, stat="probability", element="step")
```

## **5. Heatmap (Categorical - Categorical)**

```python
pd.crosstab(titanic["Pclass"], titanic["Survived"])
```

```python
sns.heatmap(pd.crosstab(titanic["Pclass"], titanic["Survived"]))
```

```python
(titanic.groupby("Pclass").mean(numeric_only=True)["Survived"]*100)
```

```python
(titanic.groupby("Pclass").mean(numeric_only=True)["Survived"]*100).plot(kind="bar")
```

```python
(titanic.groupby("Sex").mean(numeric_only=True)["Survived"]*100)
```

```python
(titanic.groupby("Embarked").mean(numeric_only=True)["Survived"]*100)
```

## **6. ClusterMap (Categorical - Categorical)**

```python
# dendograms
sns.clustermap(pd.crosstab(titanic["Parch"], titanic["Survived"]))
```

```python
flights.pivot_table(values="passengers", index="month", columns="year", observed=False)
```

```python
sns.heatmap(flights.pivot_table(values="passengers", index="month", columns="year", observed=False))
```

```python
sns.clustermap(flights.pivot_table(values="passengers", index="month", columns="year", observed=False))
```

## **7. Pairplot**

```python
# scatter plot between all numerical columns
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue="species")
```

## **8. Lineplot (Numerical - Numerical)**

```python
flights_year_wise = flights.groupby("year").sum(numeric_only=True).reset_index()
flights_year_wise
```

```python
sns.lineplot(flights_year_wise, x="year", y="passengers")
```
